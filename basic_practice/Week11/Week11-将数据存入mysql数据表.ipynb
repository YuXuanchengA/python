将练习1抽取的文章数据，以json字符串格式存入文件后。读取文件内容，将每一条记录插入事先创建好的mysql数据库中的数据表里。

```python
import json
with open("json_output.txt") as f:
    info = json.loads(f.read())
info
```

```python
import pymysql
conn = pymysql.connect('localhost','root','','kekenet')
cur = conn.cursor()
sql = 'set names utf8'
cur.execute(sql)
```

```python
for lis in info:
    title=str(lis['标题'])
    abstract=str(lis['摘要'])
    date=str(lis['日期'])
    editor=str(lis['编辑'])
    sort=str(lis['标签'])
    link=str(lis['链接'])
    sql = "insert into kekenetkeji (title, abstract, date, editor, sort, link) values ('"+title+"', '"+abstract+"', '"+date+"', '"+editor+"', '"+sort+"', '"+link+"')";
    #sql = "insert into kekenetkeji (title, abstract, date, editor, sort, link) values ('{}','{}','{}','{}')".format(title, abstract, date, editor);
    
    cur.execute(sql)
    sql = 'select * from kekenetkeji'
    cur.execute(sql)
    conn.commit()
    data = cur.fetchall()
    print('插入了一条数据:')
    print(data)
    print("\n")
```

```python
cur.close()
conn.close()
```
