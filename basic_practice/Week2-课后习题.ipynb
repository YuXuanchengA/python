```python
#计算10000以内所有2,3,4,5的倍数的正整数的和并分别打印出来

total = 0

numbers=[]

for i in range(2,10000):
    if (i%2 == 0):
        total += i
        numbers.append(i)
        i += 1
    elif (i%3 == 0):
        total += i
        numbers.append(i)
        i += 1
    elif (i%4 == 0):
        total += i
        numbers.append(i)
        i += 1
    elif (i%5 == 0):
        total += i
        numbers.append(i)
        i += 1
        
print(total)
print(numbers)
```

```python
#将可打印的ascii码全部放入一个数组中，并以每行20个的格式打印出来 

a = []

for i in range(32,127):
    i_tr = chr(i)
    a.append(i_tr)
    
for j in range(len(a)):
    print(a[j],end='')
    if(j+1)%20 == 0:
        print("\n")
```
